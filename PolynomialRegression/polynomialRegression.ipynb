Polynomial Linear Regression<br>
y = b0 + b1 * x1 + b2 * x2 ^ 2<br>
It is still called Linear instead of a x ^ 2 term in the function because it is the function of b coefficient and not x because we put the values of x in the function to predict y and for that we need to find the correct values of the coefficients.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset
```

```python
plt.scatter(dataset["Level"],dataset["Salary"],color="red")
plt.xlabel("Level")
plt.ylabel("Salary")
```

Since the above scatter plot shows that the Salary and Level are not linearly related but have a quadratic or a plynomial relation of degree 2 therefore the plynomial linear regression can be used here.

```python
x=dataset.iloc[:,1:2].values # always ensure that the is a matrix (m * n format) and to do that we need to write 1:2 
                             # and not the exact column number 1
y=dataset.iloc[:,2].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(x)
X_poly
```

```python
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
# Visualizing the linear regression results
plt.scatter(x,y,color="red")
plt.plot(x,lin_reg.predict(x),color="blue")
plt.title("Linear Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visulaizing the Plynomial Linear Regression results
X_grid=np.arange(min(x), max(x), 0.1)
X_grid=X_grid.reshape(len(X_grid),1)
plt.scatter(x,y,color='red')
plt.plot(X_grid,lin_reg2.predict(poly_reg.fit_transform(X_grid)),color='blue')
plt.title("Polynomial Linear Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# Predicting new Result with Linear Regression 
lin_reg.predict(np.array([6.5]).reshape(1,1))
```

```python
# Predicting new Results with Plynomial Linear Regression
lin_reg2.predict(poly_reg.fit_transform(np.array([6.5]).reshape(1,1)))
```

Simple linear regression

y = b0 + b1 * x

y is dependent variable 
x is independent variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,1]
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=1/3,random_state=0)
```

```python
#Linear regression does not require feature scaling because in the inbuilt model does it for us
#Fitting the simple linear regression model on the training data
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Prediciting
y_pred=regressor.predict(X_test)
y_pred
```

```python
y_test
```

```python
#Visualizing the trainging set results
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary Vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Now visualizing the test set results
#Visualizing the trainging set results
plt.scatter(X_test,y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary Vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
